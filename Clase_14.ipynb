# KNN

````python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, neighbors
from mlxtend.plotting import plot_decision_regions
def knn_comparison(data, k): # funcion de comparacion
 x = data[['X','Y']].values # Extraccion de columns
 y = data['class'].astype(int).values # Clase y como int
 clf = neighbors.KNeighborsClassifier(n_neighbors=k) #algoritmo
 clf.fit(x, y)# Graficar la region de decision
 plot_decision_regions(x, y, clf=clf, legend=2)# AÃ±adir anotaciones
 plt.xlabel('X')
 plt.ylabel('Y')
 plt.title('Knn with K='+ str(k))
 plt.show()
````

````python
# Cargar y aplicar funcion
data1 = pd.read_csv('ushape.csv')
for i in [1,5,20,30,40,80]: # Para diferentes valores de k (Knn)
    knn_comparison(data1, i)
````

````python
# Data concentrica
data2 = pd.read_csv('concertriccir2.csv')
for i in [1,5,20,30,40,60]:
    knn_comparison(data2, i)
````

````python
# Data XOR
data3 = pd.read_csv('xor.csv')
for i in [1,5,20,30,40,60]:
   knn_comparison(data3, i)
````

````python
# Linear separable
data4 = pd.read_csv('linearsep.csv')
for i in [1,5,20,30,40,60]:
    knn_comparison(data4, i)
````

## Ejercicio: Agregar datos de varias empresas

Combinemos los cinco archivos CSV para analizar las cinco empresas juntas que hamos trabajado antes. Esto tambiÃ©n reducirÃ¡ la cantidad de trabajo de programaciÃ³n requerido, ya que el cÃ³digo se compartirÃ¡ entre las cinco empresas.

Una forma de realizar esta tarea de agregaciÃ³n es utilizar el mÃ©todo ```pd.concat ()``` de ```pandas```. Una entrada en este mÃ©todo puede ser una lista de DataFrames que le gustarÃ­a concatenar. Usaremos un bucle for para recorrer cada sÃ­mbolo bursÃ¡til, cargar el archivo CSV correspondiente y luego agregar el resultado a una lista que luego se agrega usando ```pd.concat()```. Echemos un vistazo a cÃ³mo se hace esto.

````python
print("Definiendo los simbolos de stock")
symbol_data_to_load = ['D','EXC','NEE','SO','DUK']
list_of_df = []

# Ciclo sobre simbolos
#llenar la lsita de dataframes
print(" --- Inicio de Loop --- ")
for i in symbol_data_to_load:
    print("Procesando Simbolo: " + i)
    temp_df = pd.read_csv(i+'.csv',sep=',')
    temp_df['Volume_Millions'] = temp_df['Volume'] / 1000000.0
    temp_df['Symbol'] = i # Agregar nueva columna con el simbolo 
    list_of_df.append(temp_df)

print(" --- Completado loop simbolos --- ")
    
# Combinar en un Dataframe unico usando concat
#permite pegar los dataframes de la lista
print("Agregando la data")
agg_df = pd.concat(list_of_df, axis=0)

# Agregar estadisticas de retorno y volatilidad
# es mas rÃ¡pido agregarlo al dataframe que a cada uno de los registros
print('Calculando estadisticas finales')
agg_df['VolStat'] = (agg_df['High'] - agg_df['Low']) / agg_df['Open']
agg_df['Return'] = (agg_df['Close'] / agg_df['Open']) - 1.0

print("agg_df DataFrame dimension (filas, columnas): ")
print(agg_df.shape)

print("Head del DataFrame agg_df: ")
agg_df.head()

#print("agg_df['Symbol'].unique()")
````

# Modelo de regresiÃ³n (Volstat)

````python
### Load relevant packages
import pandas                  as pd
from   scipy import stats
import numpy                   as np
import matplotlib.pyplot       as plt
import seaborn                 as sns
import statsmodels.formula.api as sm
# https://community.plot.ly/t/solved-update-to-plotly-4-0-0-broke-application/26526/2
import os

#%matplotlib inline
#plt.style.use('ggplot')
from bokeh.resources import INLINE
import bokeh.io
from bokeh import *
````

Primero debemos estandarizar con el proceso min-max scaler

````python
data= agg_df[['Open','High','Low','Close','Volume_Millions','Symbol','VolStat','Return']]
data.head()
````

````python
def min_max_scaling(series):
    return (series - series.min()) / (series.max() - series.min())
for col in data.columns:
  if col == 'Symbol':
    pass
  else:
    data[col] = min_max_scaling(data[col])
data.head()
````

````python
agg_df.columns
````

# Parte 3

````python
model1 = 'VolStat~Open+ High+ Low+ Close + Volume_Millions +Symbol'
lm1   = sm.ols(formula = model1, data = data).fit()
print(lm1.summary())
````

En general el modelo tiene un alto valor de $RÂ²$ con un valor de 0.834 lo cual implica que el modelo explica el 83% de la varianza de todos los datos, de igual forma la regresiÃ³n es significativa (al menos un coeficiente es diferente de 0) del valor p obtenido de la prueba F.

Respecto a los coeficientes se puede decir que:
1. Las acciones DUK, SO y EXC presentan una menor volatilidad respecto a D mientras que NEE una mayor volatilidad
2. El coeficiente de mayor valor tiene que ver con la variable High (relacion proporcional) mientras que la variable Low tiene una correlaciÃ³n negativa con la volatilidad en ambos casos significativos por el valor p obtenido
3. El efecto comÃºn de volatilidad esperada cada dÃ­a es en promedio 0.1281
